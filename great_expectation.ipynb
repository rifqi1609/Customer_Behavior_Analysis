# Preparing Data and Tool

```python
# Create a data context
from great_expectations.data_context import FileDataContext
context = FileDataContext.create(project_root_dir='./')
```

```python
# Give a name to a Datasource. This name must be unique between Datasources.
datasource_name = 'data'
datasource = context.sources.add_pandas(datasource_name)

# Give a name to a data asset
asset_name = 'customer_dataset'
path_to_data = r'C:\Users\DELL\Portfolio\Customer_Behavior_Analysis\dags\clean_data.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
```

```python
# Creat an expectation suite
expectation_suite_name = 'expectation-customer-dataset'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name)

# Check the validator
validator.head()
```

# Setting Expectation

```python
# Expectation 1 : Column `dropoff_datetime` must be unique
validator.expect_column_values_to_be_unique('id')
```

```python
# Expectation 2 : Column `tip_amount` must be less than $ 100
validator.expect_column_values_to_be_between(
    column='age', min_value=0, max_value=100
)
```

```python
# Expectation 3 : Column `education` must contain one of the following 4 things :
validator.expect_column_values_to_be_in_set('education', ['High School', 'Undergraduate', 'Master', 'PhD'])
```

```python
# Expectation 4 : Column `income` must in form of float or integer
validator.expect_column_values_to_be_in_type_list('income', ['float', 'integer'])
```

```python
# Expectation 5 : Column 'dt_customer' should be in date data type
validator.expect_column_values_to_match_strftime_format(column='dt_customer', strftime_format='%Y-%m-%d')
```

```python
# Expectation 6 : Number of column should be 29 columns
validator.expect_table_column_count_to_equal(value=28)
```

```python
# Expectation 7 : columns' name check
validator.expect_table_columns_to_match_set(column_set=['id',
 'education',
 'marital_status',
 'income',
 'kidhome',
 'teenhome',
 'dt_customer',
 'recency',
 'mnt_wines',
 'mnt_fruits',
 'mnt_meat_products',
 'mnt_fish_products',
 'mnt_sweet_products',
 'mnt_gold_prods',
 'num_deals_purchases',
 'num_web_purchases',
 'num_catalog_purchases',
 'num_store_purchases',
 'num_web_visits_month',
 'accepted_cmp_3',
 'accepted_cmp_4',
 'accepted_cmp_5',
 'accepted_cmp_1',
 'accepted_cmp_2',
 'complain',
 'age',
 'customer_years',
 'num_no_deal_purchases'])
```

```python
# Save into Expectation Suite
validator.save_expectation_suite(discard_failed_expectations=False)
```

# Expectation Checkpoint

```python
# Create a checkpoint
checkpoint_1 = context.add_or_update_checkpoint(
    name = 'checkpoint_1',
    validator = validator,
)
```

```python
# Run a checkpoint
checkpoint_result = checkpoint_1.run()
```
